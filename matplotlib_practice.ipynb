```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
voltage = np.linspace(0, 100, 50)
resistance = 30
```

 - V = I*R

```python
current_1 = voltage / resistance
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.legend()
plt.show()
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.xlabel('Voltage')
plt.ylabel('Current')
plt.legend()
plt.show()
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.xlabel('Voltage')
plt.ylabel('Current')

plt.title("Ohm's Law")
plt.legend()

plt.show()
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.xlabel('Voltage')
plt.ylabel('Current')

plt.xlim(20, 60)
plt.ylim(1, 3)

plt.title("Ohm's Law")
plt.legend()

plt.show()
```

```python
current_2 = voltage / resistance ** 2
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.plot(voltage, current_2, label='V=I*R^2')

plt.xlabel('Voltage')
plt.ylabel('Current')

plt.legend()

plt.title("Results of an Experiment")

plt.show()
```
